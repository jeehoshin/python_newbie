## 2-3) While문

```python
i = 0
sum = 0
print('i sum')

while(sum<20):
    i=i+1
    sum=sum+1
    print(i,sum)
```

```python
while True:
    print('while test')
```

## Break과 continue

```python
k = 0
while True:
    k=k+1
    if(k>3):
        break
    print(k)
```

```python
for k in range(5):
    if(k==2):
        continue
    print(k)
```

이걸 쓰면 뛰어넘고 해주게 되는 것 같다.

```python
k = 0
while True:
    k=k+1
    if(k==2):
        print("thank you next")
    if(k>4):
        break
    print(k)
```

```python
#리스트 컴프리헨션 - 사용 안 한것 (1)
numbers = [1,2,3,4,5]
square=[]

for i in numbers:
    square.append(i**2)
    
print(square)
```

```python
#리스트 컴프리헨션 - 사용 한 것 (1)
numbers = [1,2,3,4,5]
square=[i**2 for i in numbers]
print(square)
```

```python
#리스트 컴프리헨션 - 사용 안 한것 (2)
numbers=[1,2,3,4,5]
square=[]

for i in numbers:
    if i > 3:
        square.append(i**2)
        
print(square)
```

```python
numbers=[1,2,3,4,5]
square= [i**2 for i in numbers if i>=3]
print(square)
```
