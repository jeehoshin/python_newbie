### 3) Scatter Plot

기본 syntax
```
plt.scatter(x,y [, s = size_n, c = colors, marker = 'marker_string', alpha = alpha_f])
```

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1234)
height = np.random.randint(150, 190, 100)   # 키 데이터
weight = np.random.randint(35, 100, 100)  # 몸무게 데이터

plt.scatter(height, weight)
plt.xlabel('키(m)')
plt.ylabel('몸무게(kg)')
plt.title('키와 몸무게')
plt.grid()
````

````python
plt.scatter(height, weight, s = 40, c = 'r') #마커 크기는 0.5 색깔은 red
plt.grid()
````

- 마커별로 크기와 컬러를 다르게 지정해보자

````python
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

# 위도(latitude)와 경도(longitude)
lat  = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16] 
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

# 인구 밀도(명/km^2): 2017년 통계청 자료
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den) * 0.2 # 마커의 크기 지정 
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] # 마커의 컬러 지정

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)
plt.xlabel('경도(longitude)')
plt.ylabel('위도(latitude)')
plt.title('지역별 인구 밀도(2017)')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name) # 위도 경도에 맞게 도시 이름 출력

plt.show()
````

### 4) Bar Plot

여러 항목을 비교할 때 유용한 bar_plot

```
plt.bar(x, height [, width = width_f, color = colors,  tick_label = tick_labels, align = 'center' 혹은 'edge', label = labels])
```

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#데이터 만들기
member_id = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [27,35,40,33]
after_ex = [30,38,42,37]

n_data = len(member_id)
index = np.arange(n_data)
plt.bar(index, before_ex)
plt.show()
````

- tick label 지정

````python
plt.bar(index, before_ex, tick_label = member_id)
plt.show()
````

- 컬러 지정

````python
colors = ['r', 'g', 'b', 'm']
plt.bar(index, before_ex, tick_label = member_id, color = colors)
plt.show()
````

- 폭 지정

````python
colors = ['r', 'g', 'b', 'm']
plt.bar(index, before_ex, tick_label = member_id, color = colors, width = 0.5)
plt.show()
````

- 가로 막대 그래프 그리기
```
plt.barh()
```

````python
colors = ['r', 'g', 'b', 'm']
plt.barh(index, before_ex, tick_label = member_id, color = colors) #width 옵션은 사용 불가
plt.show()
````

- 두 개의 데이터를 비교해보자

> width 옵션은 두 개의 막대 그래프가 들어갈 수 있도록 0.4정도로

> 두 데이터가 겹치지 않게 after_ex 그래프의 시작 위치는 막대 그래프의 두께(0.4)만큼 오른쪽으로 이동

> 두개의 데이터를 붙인 경우는 tick_label옵션 사용 불가. 이 경우 plt.xticks라는 함수를 사용하여 붙인다

````python
barWidth = 0.4
plt.bar(index, before_ex, color = 'k', align = 'edge', width = barWidth, label = 'before')
plt.bar(index+barWidth, after_ex, color = 'b', align = 'edge', width = barWidth, label = 'after')

plt.xticks(index+barWidth, member_id) #tick위치, 붙일 라벨
plt.legend(loc = 'upper right')
plt.xlabel('회원 ID')
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 전후 근지구력 변화')
plt.show()
````

### 5) Pie Graph

```
plt.pie(x, [, labels = label_seq, autopct = '비율 표시 형식(ex : %0.1f)', shadow = False(기본) 혹은 True,
explode = explode_seq, countreclock = True(기본) 혹은 False, startangle = 각도 (기본은 0)
```

- 기본 파이차트 그리기

파이 그래프를 그리려면 가로와 세로의 비율이 1:1이어야 한다고 함

````python
fruit = ['사과', '바나나', '딸기', '오렌지', '포도']
results = [7,6,3,2,2]

import matplotlib.pyplot as plt
plt.pie(results)
plt.show()
````

책에서는 위 그래프가 타원으로 보여야 한다는데 난 잘 보인다....

````python
plt.figure(figsize = (5,5))
plt.pie(results)
plt.show()
````

- 부채꼴 부분의 라벨 설정하기

````python
plt.figure(figsize = (5,5))
plt.pie(results, labels = fruit, autopct = '%.1f%%')
plt.show()
````

- x축 기준 90도부터 시작

````python
plt.figure(figsize = (5,5))
plt.pie(results, labels = fruit, autopct = '%.1f%%', startangle=90, counterclock = False)
plt.show()
````

- 그림자 추가와 특정 항목 강조 

````python
explode_value = (0.1, 0, 0, 0, 0) #이걸로 첫 항목만 강조를 때린다

plt.figure(figsize = (5,5))
plt.pie(results, labels = fruit, autopct = '%.1f%%', startangle=90, counterclock = False, 
       explode = explode_value, shadow = True)
plt.show()
````

    barh : 수평 바 그래프
