```python
import pandas as pd
conn = cx_Oracle.connect('hr/1234@127.0.0.1:1521/xe')
query = 'select * from departments'
df_dept = pd.read_sql(query, con = conn)
df_dept
```

```python
df_dept[['DEPARTMENT_ID', 'MANAGER_ID']]
```

### row를 기준으로 꺼내올때, 뒤에 컬럼도 선택할 수 있다.

```python
df_dept.loc[[2,3,20], ['MANAGER_ID', 'DEPARTMENT_ID']]
```

```python
df_dept.loc[[2,3,20], [3,1]] #이렇게 하려면 iloc를 사용해야 한다. 
```

```python
df_dept.iloc[[2,3,20], [3,1]] #
```

#### 이 방법을 사용하면 바로 R처럼 꺼내올 수 있는 것 같다... 드디어!!!!!

#### filter

```python
df_dept.filter(like = 'MENT', axis = 1) #axis 1은 컬럼을 기준으로 찾는다. 
```

```python
df_dept.filter(like = 'ID', axis = 1).head()
```

```python
df_dept[df_dept.DEPARTMENT_NAME == 'IT']
```

```python
df_dept[df_dept.LOCATION_ID == 1700] #이런 방식으로 디플라이어의 필터를 걸 수 있구나. 아주아주 간편...
```
